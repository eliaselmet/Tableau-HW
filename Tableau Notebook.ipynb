```python
import pandas as pd

Q12019 = pd.read_csv("2019Q1.csv")
Q22019 = pd.read_csv("2019Q2.csv")
Q12019.head()
```

```python
Q22019 = Q22019.drop(['Column 1'],axis= 1)
Q22019.head()
```

```python
YTD2019 = Q12019.append(Q22019)
June19data = pd.read_csv("201906-citibike-tripdata.csv")
June19data.head()
```

```python
June19data.dtypes
```

```python
June19data['starttime'] = pd.to_datetime(June19data['starttime'], format='%Y-%m-%d %H:%M:%S.%f').dt.strftime('%Y-%m-%d')
June19data['stoptime'] = pd.to_datetime(June19data['stoptime'], format='%Y-%m-%d %H:%M:%S.%f').dt.strftime('%Y-%m-%d')
```

```python
June19data.head()
```

```python
June0119 = June19data[(June19data['starttime'] == '2019-06-01')]
June0119 = June0119.drop(columns = ['stoptime','tripduration','bikeid','usertype','birth year','gender'])
June0119
```

```python
import numpy as np
starts = June0119['start station id'].value_counts()
starts = starts.to_frame()
starts = starts.reset_index()
starts = starts.rename(columns = {'index':'station id','start station id':'starts'})

ends = June0119['end station id'].value_counts()
ends = ends.to_frame()
ends = ends.reset_index()
ends = ends.rename(columns = {'index':'station id','end station id':'ends'})

station_lat = June0119.groupby('start station id', as_index = False)['start station latitude'].mean()
station_long = June0119.groupby('start station id', as_index = False)['start station longitude'].mean()
station_long

station_loc = station_lat.merge(station_long, on = 'start station id')
station_loc = station_loc.rename(columns = {'start station id':'station id', 'start station latitude':'lat', 'start station longitude':'long'})
station_usage = starts.merge(ends, how = 'outer', on = 'station id')
station_loc.head()
station_data = station_usage.merge(station_loc, on ='station id')
station_data = station_data.fillna(0)


station_data
```

```python
station_data.to_json('station_data.json',orient = 'table')
```
